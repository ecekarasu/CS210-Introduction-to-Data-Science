## Introduction

No-shows, the people who do not show up for their for their appointments, are a common problem in the public health sector. Decreased efficiency and reduced possibility of other patients' access to health services are two of the negative outcomes of this problem. There may be different factors that give rise to no-shows. So that, investigating no-shows considering that each individual has different characteristics is an important part for developing solutions to decrease the number of no-shows. The goal of this project is to find correlations between the information of whether s/he showed up for the appointment and multiple features of that person, such as the age, gender, neighbourhood... By doing this kind of analysing, it is expected to come to a conclusion about  people who are more likely to not to show up for the appointment.

### Problem Definition

The problem stated in this project is the patients who do not show up for their scheduled appointments. They are considered as a problem because as the number of no-shows increases, the efficiency of both the appointment system and the doctors decreases. The end goal of this project is to avoid this problem by predicting as many people as possible who will not come to the appointment. In order to achieve this, data containing the information of a group of people will be examined. 
According to the results drawn from this data, having evidence-based predictions will be possible.

### Utilized Datasets

In this project, the dataset is taken from https://www.kaggle.com/joniarroba/noshowappointments. The dataset consists of 110.527 rows, standing for the appointments, and 14 columns, each of them are associated with different features of the people and appointments. 

In detail; the information kept in the columns are patient id, appointment id, gender, schedule day, appointment day, age, neighbourhood, scholarship, hipertension, diabetes, alcoholism, handcap, sms received, no-show in that order.

Schedule day is when someone called or registered the appointment and appointment day is the actual day of the appointment when the patient comes and visits the doctor. Appointment id and neighbourhood are two other columns related to the appointment. They are used for the identification of each appointment and 
specification of the location of the appointment. Scholarship, hipertension, diabetes, alcoholism, handcap, no-show are binary variables that change depending on whether they exist in the patient. 

According to these variables, two types of data were used, numerical and categorical. The binary variables make up the data used categorically (two categories: true, false), with the columns gender and neighbourhood. The others are considered as the data used numerically.

## Data Exploration

In the initial data exploration phase, the structure and columns of the data were thoroughly analyzed to identify any missing values or outliers. Upon examination, it was determined that the dataset did not contain any missing entries or duplicates, so no action was taken to address these issues. Outliers in the Age column were identified using the interquartile range, and any values below the lower bound or above the upper bound were removed. The data types of the columns were also evaluated, and it was determined that the ScheduledDay and AppointmentDay columns should be converted to datetime data types due to their default object data type being incorrect. In addition, new columns were created for the month, day, and hour of the appointments to facilitate grouping and analysis, as these factors may have a significant impact on the no-show rate for medical appointments.

In addition to the previously mentioned steps, encoding was also applied to the categorical columns named Neighborhood, No show, and Gender. This process involved converting the categorical data into numerical values, which can be more easily analyzed and used in machine learning models. By encoding the data in this way, it is possible to more accurately identify patterns and trends within the data, and to more effectively use this information in predictive modeling efforts.

To decipher "the Medical No Shows", the graphics obtained from the main heading columns in the dataframe are used in such a way that presenting the material that will support the understanding of the relationships between the features was possible.
